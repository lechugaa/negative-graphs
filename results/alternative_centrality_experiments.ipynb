# Centrality Metrics Experiments

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import seaborn as sns


sns.set_context('talk', font_scale=1.2)
plt.rcParams["font.family"] = "Times New Roman"

%matplotlib inline
```

## Barabasi variación del tamaño del grafo

* ACM_001.csv – BA M=20 N=[50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

```python
ba_node = pd.read_csv('alternative_centrality_metrics/ACM_001.csv')
ba_node.head()
```

### Incertidumbre vs fracción

```python
data = ba_node

g = sns.relplot(x='real_fraction',
            y='graph_uncertainty', 
            hue='graph_size', 
            data=data,
            height=10, aspect=1, s=100)

g.set(xlabel='Fake-to-real edge proportion', ylabel='Total uncertainty [bits]')
g._legend.set_title('Size')
for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```

```python
data = ba_node[ba_node.graph_size <= 250]

g = sns.relplot(x='real_fraction',
            y='graph_uncertainty', 
            hue='graph_size', 
            data=data,
            height=10, aspect=1, s=100)

g.set(xlabel='Fake-to-real edge proportion', ylabel='Total uncertainty [bits]')
g._legend.set_title('Size')
for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```

### Error cudrático medio betweenness

```python
metric = 'betweenness'
data = ba_node[ba_node.centrality_metric == f'{metric}_centrality']

g = sns.relplot(x='mean_uncertainty',
                y='mean_se_value', 
                hue='graph_size', 
                data=data[data.graph_size >= 0], 
                height=10, aspect=1, s=100)

g.set(xlabel='Mean uncertainty [bits]', ylabel='Betweenness centrality MSD')
g._legend.set_title('Size')

for ax in g.axes.flat:
    ax.yaxis.set_major_formatter(tkr.FuncFormatter(lambda y, p: f'{y * 10000:.1f}' + 'x10$^{-4}$'))
```
