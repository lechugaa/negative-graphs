# Centrality Metrics Experiments

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import seaborn as sns


sns.set_context('talk')

%matplotlib inline
```

## Barabasi variación del tamaño del grafo

* ACM_001.csv – BA M=20 N=[50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

```python
ba_node = pd.read_csv('alternative_centrality_metrics/ACM_001.csv')
ba_node.head()
```

### Incertidumbre vs fracción

```python
data = ba_node

g = sns.relplot(x='real_fraction',
            y='graph_uncertainty', 
            hue='graph_size', 
            data=data)

g.set(xlabel='Fracción', ylabel='Incertidumbre [bits]')
g.fig.suptitle('Incertidubre total vs fracción')
g._legend.set_title('Nodos')
plt.subplots_adjust(top=0.90)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```

```python
data = ba_node[ba_node.graph_size <= 250]

g = sns.relplot(x='real_fraction',
            y='graph_uncertainty', 
            hue='graph_size', 
            data=data)

g.set(xlabel='Fracción', ylabel='Incertidumbre [bits]')
g.fig.suptitle('Incertidubre total vs fracción')
g._legend.set_title('Nodos')
plt.subplots_adjust(top=0.90)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```
