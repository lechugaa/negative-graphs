# Centrality Metrics Experiments

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import seaborn as sns


sns.set_context('talk')

%matplotlib inline
```

## Erdos-Renyi variación del tamaño del grafo

* CM_001.csv – ER P=0.25 N=[50, 100, 150, 200, 250, 300, 350, 400, 450, 500]

```python
er_node = pd.read_csv('centrality_metrics/CM_001.csv')
er_node.head()
```

### Incertidumbre vs fracción

```python
data = er_node[er_node.graph_size >= 250]

g = sns.relplot(x='fraction',
            y='graph_uncertainty', 
            hue='graph_size', 
            data=data)

g.set(xlabel='Fracción', ylabel='Incertidumbre [bits]')
g.fig.suptitle('Incertidubre total vs fracción')
g._legend.set_title('Nodos')
plt.subplots_adjust(top=0.90)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```

### Incertidumbre promedio vs fracción

```python
data = er_node[er_node.graph_size >= 250]

g = sns.relplot(x='fraction',
            y='mean_uncertainty', 
            hue='graph_size', 
            data=data)

g.set(xlabel='Fracción', ylabel='Incertidumbre promedio [bits]')
g.fig.suptitle('Incertidubre promedio vs fracción')
g._legend.set_title('Nodos')
plt.subplots_adjust(top=0.90)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```

### Error cuadrático medio de centralidad de grado

```python
data = er_node[er_node.centrality_metric == 'degree_centrality']

g = sns.relplot(x='mean_uncertainty',
            y='mean_se_value', 
            hue='graph_size', 
            data=data[data.graph_size >= 250])

g.set(xlabel='Incertidumbre promedio [bits]', ylabel='Error cuadrático medio')
g.fig.suptitle('ECM centralidad de grado vs incertidumbre promedio')
g._legend.set_title('Nodos')
plt.subplots_adjust(top=0.90)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.2f}'))
    ax.get_xaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```

### Error cuadrático medio de closeness

```python
data = er_node[er_node.centrality_metric == 'closeness_centrality']

g = sns.relplot(x='mean_uncertainty',
            y='mean_se_value', 
            hue='graph_size', 
            data=data[data.graph_size >= 250])

g.set(xlabel='Incertidumbre promedio [bits]', ylabel='Error cuadrático medio')
g.fig.suptitle('Closeness vs incertidumbre promedio')
g._legend.set_title('Nodos')
plt.subplots_adjust(top=0.90)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.2f}'))
    ax.get_xaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```
