```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import seaborn as sns


sns.set_context('talk')
```

```python
ct_df = pd.read_csv('contact_tracing_5000.csv')
ct_df.head()
```

```python
ct_df.groupby(['fraction'], as_index=False).mean()
```

```python
data = ct_df.groupby(['fraction'], as_index=False).mean()

g = sns.relplot(x='fraction',
            y='number_contacts', 
            data=data)

g.set(xlabel='Fracción', ylabel='Número de contactos', ylim=(0,2000))
g.fig.suptitle('Número de contactos vs fracción')
plt.subplots_adjust(top=0.90)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```

```python
data = ct_df.groupby(['fraction'], as_index=False).mean()

g = sns.relplot(x='mean_uncertainty',
            y='number_contacts', 
            data=data)

g.set(xlabel='Incertidumbre promedio [bits]', ylabel='Número de contactos', ylim=(0,2000))
g.fig.suptitle('Número de contactos vs incertidumbre promedio de nodo')
plt.subplots_adjust(top=0.85)

for ax in g.axes[:,0]:
    ax.get_yaxis().set_major_formatter(tkr.StrMethodFormatter('{x:,.0f}'))
```
